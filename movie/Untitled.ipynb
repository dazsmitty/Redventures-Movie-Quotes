```python
file1 = open("moives.txt","r") 
```

```python
movies=file1.readlines()
```

```python
not_found
for i in range(len(movies)):
    movies[i]=movies[i].strip
    movie=movies[i].replace(" ","_")
    page_link="https://www.rottentomatoes.com/m/"+movie
    page_response = requests.get(page_link, timeout=5)
    page_content = BeautifulSoup(page_response.content, "html.parser")
    mydivs = page_content.findAll("section", {"class": "panel panel-rt panel-box","id":"quotesList"})
    result = []
    for th in mydivs:
        result.extend(th.find_all(text='There are no approved quotes yet for this movie.'))
    if len(result)==0:
        not_found.append(movie)
    else:
        page_link=page_link+"/quotes/"
```
